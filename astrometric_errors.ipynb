```python
import numpy as np
import matplotlib.pyplot as plt
#import seaborn as sns
from sqlalchemy import create_engine
import pandas as pd
from astropy.table import Table
from astropy.io import fits
from astropy.io import ascii
from astropy.nddata import Cutout2D
import gc

%matplotlib inline
```

```python
from properimage import utils
```

```python
engine = create_engine('sqlite:///../RBpipeline/rbogus-dev.db')
simulated = pd.read_sql_query("""SELECT * FROM Simulated""", engine)
sim_images = pd.read_sql_table('Images', engine)
```

```python
len(simulated)
```

```python
sims = []
detc = []
ids = []
for sim_image in sim_images.itertuples():
    cat = simulated[simulated['simage_id']==int(sim_image.id)]
    cat = Table.from_pandas(cat)
    imgpath = sim_image.path.strip('diff.fits')+'s_corr_detected.csv'
    
    detected = ascii.read(imgpath, format='csv')
    detected['x']=detected['X_IMAGE']
    detected['y']=detected['Y_IMAGE']
    
    ids.append(utils.matching(cat, detected, radius=4, masked=True))
    
    sims.append(cat)
    detc.append(detected)
```

```python
total=0
for _, mask in ids:
    total +=sum(mask)
print total
```

```python
plt.figure(figsize=(18, 18))
for cat in sims:
    plt.plot(cat['x'], cat['y'], 'rx')
for i, detected in enumerate(detc):
    ids_, mask = ids[i]
    #print detected['x'][mask]
    plt.plot(detected['x'][mask], 
             detected['y'][mask], 'bx')
plt.xlim(-10, 1010.)
plt.ylim(-10, 1010.)
```
